# Model Building

Mathematically cosine similarity is the best techinique for Semantic similariy analysis.

Here we are going to match our products by using cosine similarity opeartions.

>> The first step of the process is converting the text into embedding vectors.
>> Then applying cosine similarity mathematical operation for the products.
>> It will give scores between 0 to 1.0

```python
import transformers
from transformers import pipeline
```

```python
amazon = ama[['product_name_amazon','retail_price_amazon','discounted_price_amazon']]
flipkart = flip[['product_name','retail_price','discounted_price']]
```

```python
from sentence_transformers import SentenceTransformer, util
```

```python
import torch
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Embeddings for amazon and flipkarts products

am_embed = model.encode(amazon['product_name_amazon'], convert_to_tensor=True)
Flip_embed = model.encode(flipkart['product_name'], convert_to_tensor=True)
```

```python
## Applying cosine similarity's

cosine_scores = util.cos_sim(am_embed, Flip_embed)
```

```python
# Taking flipkarts products and comparing with amazon of each product
similar_products = []

for i in range(2000):
    maximum_score = torch.topk(cosine_scores[i], k=5, largest=True, sorted=True).indices
    for ind in maximum_score:
        if ind !=i:
            similar_products.append(ind)
            break
```

```python
similar_products = [i.item() for i in similar_products]
```
